# Differences between shared methods

```python
# A 1-D Series will have different method behavior than a 2-D Dataframe
```

```python
revenue = pd.read_csv("revenue.csv", index_col="Date")
revenue
```

```python
# Summing a Series will add down the single column it represents
series_sample = pd.Series([1, 2, 3])
series_sample.sum()
```

```python
# Summing a Dataframe will also add down the columns by default
# Same as revenue.sum(axis="index") or revenue.sum(axis=0)
revenue.sum()
```

```python
# To add across the row instead, do this
revenue.sum(axis="columns")
revenue.sum(axis=1)
```

# Quick recap on value_counts method

```python
nba = pd.read_csv("nba.csv")
nba.head()
```

```python
# At the very end, the number "1" denotes the number of rows with the exact same data (duplicates)
nba.value_counts()
```

```python
# But it can also be used to count the number of times a value appears in a certain column
# I normalized them here, so converted the #s to % of total
nba["Height"].value_counts(normalize=True).sort_index() * 100
```

# Fill in NaNs with the fillna Method

```python
# Universal fallback value
# Creates a VIEW
nba.fillna(0)
```

```python
# Creates a VIEW
nba["College"].fillna("Unknown")
```

```python
# Mutate the actual Dataframe with inplace=True
nba["College"].fillna("Unknown", inplace=True)
nba
```

```python
# The better syntax, since inplace is deprecated
nba = pd.read_csv("nba.csv")
nba["College"] = nba["College"].fillna("Unknown Test")
nba
```

# The astype Method 1

```python
# Best practice is to fill NaNs on import where needed
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0)
nba.tail()
```

# The astype Method 2 (Categories)

### Useful for when you have a small number of unique values in a column, relative to the size of the DF
### Columns that would hold a small number of possibilities, ie Months, severities, VPRs, checktype, etc.
#### Under the hood: rather than creating new copy for each row, one copy of each possibility is created then pointed to by cells in the DF

```python
nba.tail()
```

```python
nba.dtypes
```

```python
len(nba)
```

```python
# Check for unique values
nba["Position"].nunique()
```

```python
# Look, you can see the number of non-null values AND dtypes in one view! AND MEM USE!!! How cool is that?!
nba.info()
```

```python
# Create a category for a column
# Note the Categories line on the output
nba["Position"] = nba["Position"].astype("category")
```

```python
# Notice the reduced memory footprint and new dtype
nba.info()
```

```python
nba["Team"].nunique()
```

```python
# Even further reduced! Now over 10% off of footprint
nba["Team"] = nba["Team"].astype("category")
nba.info()
```

# Sort a Dataframe with sort_values Method 1

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
nba["Name"].sort_values()
```

```python
# Sort values based on a column, default ascending
# Does not mutate original dataframe
nba.sort_values(by="Name")
```

```python
nba.sort_values("Age")
```

```python
# Put NaNs at top, then descinding for the rest
nba.sort_values("Salary", na_position='first', ascending=False)
```

```python
nba = nba.sort_values("Salary", na_position='first', ascending=False)
nba
```

# Sorting on multiple columns

```python
# Columns will sort in the order they are specified
# IE Teams will be sorted, then Names will be sorted within the teams
nba.sort_values(by = ["Team", "Name"])
```

```python
nba.sort_values(by=["Age", "Weight"], ascending=False)
```

```python
# To perform different sort orders, use a list again
nba.sort_values(by=["Age", "Weight"], ascending=[False, True])
```

# Rank Method

```python
nba = pd.read_csv("nba.csv")
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
nba.head()
```

```python
# Where does this value fall within the set if it were sorted, without actually sorting it
nba["Salary"].rank()
nba["Salary"].rank(ascending=False)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype(int)
nba
```

```python
nba.sort_values("Salary", ascending=False)
```
