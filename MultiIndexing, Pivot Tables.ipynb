# Chane Index Level Names with the `set_names` Method

```python
# So, you can set multi-index on initialization
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.index = bigmac.index.set_names(names = ["Day", "Location"])
bigmac
```

```python
bigmac.index = bigmac.index.set_names(names = "Date", level=0)
bigmac
```

```python
bigmac.index = bigmac.index.set_names(names="Country", level="Location")
bigmac
```

# Extract Rows from MultiIndex DataFrame

```python
# So, you can set multi-index on initialization
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.loc["2010-01-01"].head()
```

```python
# You can do this, but it's ambiguous, cause the second key is usually referencing a column
bigmac.loc["2010-01-01", "Argentina"]
```

```python
# See? Different dtype if you select price instead of country
bigmac.loc["2010-01-01", "Price in US Dollars"]
```

#### Instead do this

```python
# The first argument is a tuple of indices, the second argument is a column name
bigmac.loc[("2010-01-01", "Argentina"), "Price in US Dollars"]
```

```python
# Using .iloc[]
bigmac.head(3)
```

```python
# iloc is looking for the ROW at index position n
bigmac.iloc[0]
```

```python
bigmac.iloc[[0, 2, 12, 15]]
```

## The `.transpose()` Method

```python
# Notice that we now have a MULTI-LEVEL-COLUMN dataframe
bigmac = bigmac.transpose()
bigmac
```

```python
bigmac.loc[("Price in US Dollars",), ("2010-01-01",)]
```

```python
bigmac.loc[("Price in US Dollars",), ("2010-01-01","Sri Lanka"):("2010-01-01", "Ukraine")]
```

### Another way to look at MultiIndex is this:


```python
wstats = pd.read_csv("worldstats.csv", index_col=("country", "year"))
wstats
```

```python
stats = {"Arab World": 
            {2015: 
                {"Population": 392, "GDP": 2.53}}}

stats["Arab World"][2015]["GDP"]
```

## Using the `.stack()` Method

```python
# Basically, we shifted columns into rows.
# So now instead of Population and GDP columns, you have a separate row for each.
# The column is unlabelled in this instance
# This is a 3-level Series now lol
world = wstats.stack()
world
```

```python
# Convert it back to a dataframe object, but it will not recover original columns
world.to_frame()
```

# The `.unstack()` Method, Part 1

```python
world
```

```python
# By default, moves the inner most index to columns
world.unstack()
```

```python
# Notice that the next index to unstack goes to the inner level columns
world.unstack().unstack()
```

```python
# Once we run out of index labels, the columns become the index labels in a series
world.unstack().unstack().unstack()
```

```python
world.unstack().unstack().unstack().to_frame()
```

# The `.pivot` Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales
```

```python
sales.pivot(index="Date", columns="Salesman", values="Revenue")
```

# The `.pivot_table()` Method

```python
foods = pd.read_csv("foods.csv")
foods.head(3)
```

```python
# In a pivot table, values should usually be numbers
# Aggfunc performs the operation on values
foods.pivot_table(values="Spend", index="Gender", aggfunc="mean")
```

```python
foods.pivot_table(values="Spend", index="Item", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], columns="City", aggfunc="sum")
```

# The `pd.melt()` Method
Creates a more tabular data structure for processing
Basically, if you have multiple column holding the same kind of datapoint - ie revenue - then it can be compressed into a single column

```python
sales = pd.read_csv("quarters.csv")
sales
```

```python
# The id_vars param determines which column will be "preserved"
pd.melt(sales, id_vars="Salesman").head(10)
```

```python
# Set column names w/ var_name and value_name
pd.melt(sales, id_vars="Salesman", var_name="Quarter", value_name="Revenue").head(10)
```
