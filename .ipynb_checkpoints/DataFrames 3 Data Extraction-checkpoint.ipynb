# Retrieve Rows by Index Label with .loc[] Accessor

```python
# Note Index = Film
bond.head()
```

```python
# Step 1 is to sort the values
bond.sort_index(inplace=True)
bond.head()
```

```python
# If only one matching row, you get a Series
bond.loc["Goldfinger"]
```

```python
# If multiple matching rows, you get a filtered dataframe
bond.loc["Casino Royale"]
```

```python
# Splice dataframe on index labels (dependent on order)
bond.loc["Diamonds Are Forever":"From Russia With Love"]
```

```python
# Also supports step splicing, or any other splicing operation you can do on a list
bond.loc["Diamonds Are Forever": "From Russia With Love": 2]
```

```python
# Looking for multiple index labels
# Will return in provided order
bond.loc[["Die Another Day", "Octopussy"]]
```

# Rename Index Labels or Columns in a Dataframe

### Use mapper and axis

```python
# Use Dictionary to map values
# Does not mutate original dataframe
index_map = {"GoldenEye": "Golden Eye"}
cols_map = {"Actor": "Actress"}

# Rename index values
bond.rename(mapper=index_map)

# Second method to rename index values
bond.rename(index=index_map).loc["Golden Eye"]
```

```python
# Rename column headers
bond.rename(mapper=cols_map, axis=1)

# Second method to rename column headers
bond.rename(columns=cols_map).head(1)
```

```python
# Third method to rename column headers
bond.columns = ['Release Year', 'Actor', 'Director', 'Box Office', 'Cost',
       'Actor Salary']
bond.head(1)
```

# Using the .nsmallest() and .nlargest() Methods

```python
# Long syntax for top 3 grossing films
bond.sort_values("Box Office", ascending=False).head(3)
```

```python
# Short syntax for top 3 grossing films
# Also much, much faster on the backend apparently
bond.nlargest(3, columns="Box Office")
```

```python
# Lowest grossing films
# Note that it auto sorts for you
bond.nsmallest(5, columns="Box Office")
```

```python
# Grab just the Series
bond["Box Office"].nlargest(5)
```

```python
bond["Box Office"].nsmallest(5)
```

# Filtering with the where Method

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask]
```

```python
# Replaces non-matching rows with null values, but keeps the index column
# More of a visual aide than a practical thing it would seem
bond.where(mask)
```

```python
bond.where(bond["Box Office"] > 800).fillna("")
```

```python
mask2 = bond["Box Office"] > 800
bond.where(mask & mask2).fillna("")
```

# The .query() Method

## This method ONLY works if there is NO whitespace in column names

```python
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
bond.head(1)
```

```python
# Apparently faster on larger datasets than using full syntax
bond.query('Actor == "Sean Connery"')
```

```python
bond.query("Director == 'Terence Young'")
```

```python
bond.query("Actor.str.contains('Conn')")
```

```python
bond.query("Actor == 'Roger Moore' and Director == 'John Glen'")
```

```python
bond.query("Actor in ['Timothy Dalton', 'George Lzenby']")
```

# Review on .apply() Method on Single Columns

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def do_thing(number):
    return str(number) + " MILLIONS!"

bond["Box Office"].apply(do_thing)
```

```python
bond.head(3)
```

```python
bond["Box Office"].apply(lambda x: str(x) + " MILLION!!" if x > 500.0 else x)
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(do_thing)
```

```python
bond.head()
```

# Use .apply() on row values

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movies(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "Awesome"
    elif actor == "Roger Moore" and budget > 40:
        return "OK"
    else:
        return "wet fart"

# Its weird, but you use axis=columns for rows lol
bond.apply(good_movies, axis="columns")
```
