# **Why does dating suck so much?**

## **Preface:**
I will start by saying this - I think that dating in general is objectively more difficult today than it has been for generations past. I think it is more difficult for both men and women, but for different reasons. The main drivers of this difficulty are the accelerating cultural and technological changes seen in Western society over the past century or so.

## Part 1: Technological Changes
Most people I see complaining about the difficulties of dating are in their 30s or younger. Which is to be expected. But the difference between today's young adults and yesterday's is that today, the majority of our lives and social interactions take place online. We share our lives through instagram posts, tiktoks, and tweets. We have ready access to the internet in our pocket at all times, expanding our social circles well past what was possible just 20 years ago. We can talk to someone 3000 miles away almost as easily as talking to someone right next to you. We regularly interact with people much older and younger than us here on reddit, with varied perspectives, without thinking much about it. There is an inherent expectation that you have your phone near you, and can respond in a timely manner. Hell, even our language has changed to [better convey emotion through text](https://www.amazon.com/Emoji-Code-Linguistics-Behind-Scaredy/dp/1250129060/ref=sr_1_1?ie=UTF8&qid=1502485655&sr=8-1&keywords=The+Emoji+Code%3A+The+Linguistics+Behind+Smiley+Faces+and+Scaredy+Cats&tag=nypost-20&asc_refurl=https://nypost.com/2017/08/12/emojis-actually-make-our-language-way-better/&asc_source=web). 
This has fundamentally changed the way in which we interact with one another, for better and for worse. Obviously, I'm here to talk about how it's made things worse. [There](https://journals.sagepub.com/doi/abs/10.1177/0894439318793947) [have](https://www.cureus.com/articles/31508-social-media-use-and-its-connection-to-mental-health) [been](https://www.emerald.com/insight/content/doi/10.1108/MHSI-06-2020-0039/full/html) [countless](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0231924) [studies](https://scholarcommons.scu.edu/engl_176/2/) examining the social and psychological impacts of social media. We know that young people today feel more alone than ever before. We know that chasing attention in the form of likes and shares is addicting, and can consume people's lives. We know that people tend to project the best possible image of themselves (from the perspective of their in-group) online, hiding their flaws and insecurities. We know that young people today are isolating themselves more than ever before, simply because it is possible (and easy) to live your life online.Now, bring an app like Tinder into the scene and the problems just compound. We gravitate towards using these apps, because we have constant easy access to them. We try so hard to make ourselves look like a good prospect. We ghost people, ignore people, and judge people based on the first few interactions we have, rather than take the time to build a relationship with whoever we first match with. We get [choice paralysis](https://www.scienceofpeople.com/choice-paralysis/), where even when we can tell that someone is a good match, we don't put the proper effort in, because what if there's an even better match right around the corner? A single bad picture, a single goof, a single faux paus, is enough to turn people off. This isn't even getting into [Tinder economics](https://theblog.okcupid.com/a-womans-advantage-82d5074dde2d).

## Part 2: Predatory Industry
Even if you do not use dating apps, you're still going to have a harder time than generations past. Studies show that in 2017, [39% of adult relationships in the United States began online](https://news.stanford.edu/2019/08/21/online-dating-popular-way-u-s-couples-meet/). Today, I would expect that to be at least 50%. That means up to half of the dating pool is caught in the dating app trap, limiting their ability to form a meaningful relationship with you. That means that the predatory dating industry is effecting your chances at finding love, too.
[Match Group essentially has a monopoly in the online dating market](https://www.vox.com/the-goods/2019/2/11/18220425/hinge-explained-match-group-tinder-dating-apps), and they have no incentive to make their algorithms better. None. In fact, they [buy out](https://www.prnewswire.com/news-releases/iacs-matchcom-acquires-okcupid-115090674.html) other companies that do have better algorithms and turn them into tinder clones. [A recent example would be Hinge](https://www.theverge.com/tech/2018/6/20/17485750/match-group-acquire-hinge-deal-online-dating). It's algorithm actually works surprisingly well for matching you with like-minded people -- ***for now***. 5 years from now, I can practically guarantee that will no longer be the case.
Dating apps make the playing field as unlevel as they can, while still having plausible deniability that they're doing so. This imbalance happens slowly over the course of years. As it does, they offer increased levels of purchases and subscriptions to level it back out (for those who cough up).  Does anyone remember how easy Tinder was to use back in 2015? It was a completely different ballgame. Now, the app is designed much like a slot machine. You're always *so close*, yet never seem to get where you want to be. For most guys, *most of the time* it feels hopeless, then suddenly you get a few matches, reinvigorating you and encouraging you to just keep on trying. Some guys will just pay up to get that rush of validation, and that is what drives the entire business model. This imbalance is often strongly in favor of straight women, who will accumulate matches much faster than most men could ever dream of. 
However, getting lots and lots of matches is actually a bad thing - for everyone. 
**Ever wonder why tinder is the way it is?**
Remember when I mentioned [choice paralysis](https://www.scienceofpeople.com/choice-paralysis/) before? This is the biggest problem that dating apps cause for women (although it is definitely a thing for some guys too). A lot of women will have their pick of dozens, maybe hundreds of men. Not through any fault of their own, but simply because of how these apps are designed to work. Then the question becomes, how are you supposed to tell if someone is compatible when the only thing you have to go off of is a few pictures and a short bio? Well, you set some standards and choose a few attractive guys that check your boxes. This selectivity naturally leads to less attractive guys (physical or otherwise) getting less attention. Then, there is always that prospect of someone *better* being just around the corner, discouraging you from getting too invested in anyone you match with, because swiping on some more people is as easy as opening the app on that phone you have on you 24/7. And the more people you match with, the more conversations you feel obligated to hold, the more likely you are to get overwhelmed, the more likely you are to put less effort into most conversations. Consequently, this leads to a smaller group of men receiving more attention than the rest, which lands them in the same position you were just in, making them now less likely to give you attention. Which leads to smaller group of women receiving even more attention than the rest, and the cycle continues.
What if instead of presenting you with near-unlimited options, a dating app did this: after learning your preferences and placing you in the correct buckets with like-minded people, you are delivered 3 potential partners once per week, and that's it. No swiping, no seeing other people's profiles. This may sound restricting, and it is, but this eliminates 99% of that choice paralysis. It encourages you to interact with the people you've matched with, because you're not going to get another set of people to talk to until next week. This, in turn, encourages you to try harder to make things work, to be more invested. That, in turn, leads to more engaged conversations. And that leads to a higher feeling of satisfaction, and finally to a higher chance of forming a meaningful relationship.
Thing is, that is not profitable. What would be more profitable is to do the above, but allow girls to have more matches than guys. Then to allow guys to pay to have more matches more often. Then let people pay to be delivered to more people. Then tune the algorithm to deliver kind-of-but-not-really compatible people so users are less likely to leave the app in pairs. And that, my friends, is why the dating app industry is predatory as hell. Because it's just too damn easy to profit off of. The only way to find success is to not play their game. Have some self control. Only swipe on people you actually want to talk to. Stop swiping and remove yourself from the deck once you have a match or two. Take the time to get to know this person, like you would someone you met in real life. Be polite, don't assume the worst. 

## Part 3: Cultural Changes
In the not-so-long-ago past, cultural norms in the Western world were quite different than what they are today. Men have, throughout most of human history (with a few interesting exceptions), been the pursuers and the deciders when it comes to relationships. The concept of dating that we recognize came about in the early 20th century as technology allowed us to travel further and communicate more often with people outside of our immediate area. This, along with other forces such as expanded women's rights, increasing standards of living, and a rapidly booming population meant that people had more choice than ever before in who they could settle down with. This led to the normalization of dating as a way of getting to know someone with romantic intent. It meant men now had to approach the women they were interested in and court them while the women would try to attract the attention of the men they were interested in. 
The concept of free love took off in the 60s. The idea that a wife's job was to serve her husband became less and less of a norm. Women increasingly became independant and joined the workforce. Men increasingly learned to be more respectful towards women. Companies like Disney shaped our ideas of love and commitment. The idea of a "soul mate" came around. Religion's grip on society loosened. But at the same time, the beauty industry rose, standards for men and women were pushed onto us, the importance of being sexy was drilled into our brains, and advertisers got all up in our heads in so many ways I can't even begin.
In just the last ~30 years alone, the internet rose to prominence, cell phones came around, then smart phones, then social media, then dating apps. Then a massive cultural shift towards equality, mutual respect, and political correctness. The #MeToo movement brought sexual assault to the forefront of public conciousness. Men were increasingly called out for coming on too strong, forcing things, flirting too much, etc. What makes a man a man came into question. Women became far more sexually liberated. The list goes on.
So many changes have happened, so many previously accepted norms have been called into question, that everyone seems to have their own idea of what courting should look like and the roles that men and women should play. This has really muddied the waters when it comes to dating. How to court a woman without going too far seems to me to be the crux of the issue for most young men. Although some things are obvious, the lines of what is and is not acceptable behavior are unclear. 
Is it okay send a flirty text? Is it okay to approach? Am I coming off as creepy? Am I coming on too strong? Do I insist that I pay the bill? Do I go in for a kiss, or is that too far? Is it okay to touch her? Is it okay to make a sexual joke? Do I initiate or wait for her to initiate? Is she in bed with me because she wants to be, or because she feels obligated? 
For a lot of guys, the last thing they want to do is cross a line and blow it. Or worse, make her uncomfortable and get labeled a creep. For guys who care about doing the right thing, it can be anxiety inducing. For guys with anxiety, it can be paralyzing. Which is bad news, because there is [an anxiety epidemic going on](https://www.washingtonpost.com/business/an-epidemic-of-depression-and-anxiety-among-young-adults/2020/08/22/dcc06c34-e43c-11ea-82d8-5e55d47e90ca_story.html). This risk-averse approach to dating causes us to overthink and play it safe, which leads to less success and less interactions overall. For guys who don't give a shit, it doesn't matter. Hence why it can appear like "girls only go for bad guys." It's just a numbers game to them.
For girls, the questions become confusing as well. Do guys like when girls approach? How direct is too direct? If I approach men, will people think I'm loose? How do I ensure I'm safe? How do I know I can trust this guy? Is he just telling me what I want to hear? Do I let that joke slide, or is that a red flag? Is it okay to kiss on the first date? How do I send a clear signal? When is it acceptable to sleep with a guy without being looked down on for it? Am I obligated to put out if he paid for dinner?  
These are questions that only continue to get more confusing as time goes on. Answers that seemed acceptable 10 years ago, may not seem that way today. Ever gone back and watched an old show and felt uneasy at some of the jokes and tropes? Yeah, that's the same effect at work here. 
If you seek out answers to these questions, you will receive a million different responses. Some will say stop thinking so much and just go for it. Some will say don't do that, it's creepy. Some will say sleeping on the first date is slutty. Some will say there's nothing wrong with showing off on OnlyFans. Some will teach you to be a [Pick Up Artist](https://www.youtube.com/watch?v=7wH2YySpoXE). Some will tell you to be yourself. Some will try to sell you a magazine subscription. With all of these conflicting answers, the whole thing can feel overwhelming and eventually not worth the effort it appears to take.
Then societies began to fracture into smaller in-groups as culture wars between those who resisted these changes and those who drove them intensified. Then COVID happened, and more fractures appeared. The dating pool contracted for all of us, as we chose sides and drew lines in the sand on who we are willing to interact with. Work from home, school from home, everything at home moved the majority of our interactions online, which has been great for Match Group and awful for the rest of us.
